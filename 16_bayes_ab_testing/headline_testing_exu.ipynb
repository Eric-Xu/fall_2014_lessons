```python
import pandas as pd
import numpy as np
from scipy.stats import beta, norm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot_beta(views, ctr):
    upper = beta.ppf(1.0-0.0001, views*ctr, views*(1-ctr))
    xs = np.linspace(0,upper,2000)
    ys = [beta.pdf(x, views*ctr, views*(1-ctr)) for x in xs]
    ys /= (sum(ys) / ((1.0*len(ys))))
    plt.plot(xs,ys)
```

```python
def demonstrate(a_views, b_views, a_ctr, b_ctr, size):
    pb = percent_better(a_views, b_views, a_ctr, b_ctr, size)
    
    if a_ctr > b_ctr:
        winner = 'A'
    else:
        winner = 'B'
        pb = 1.0 - pb
    
    plot_beta(a_views, a_ctr)
    plt.hold(True)
    plot_beta(b_views, b_ctr)
    plt.legend(['Headline A (views: %s, CTR: %0.2f)'% (a_views, a_ctr),'Headline B (views: %s, CTR: %0.2f)' % (b_views, b_ctr)]);
    plt.title('Headline %s is better in %.2f%% of the cases' % (winner, pb*100.0));
    plt.xlabel('CTR')
    plt.xlim(0,0.25)


    
def percent_better(a_views, b_views, a_ctr, b_ctr, size):
    ra = beta.rvs(a_views*a_ctr, a_views*(1-a_ctr), size=(size))
    rb = beta.rvs(b_views*b_ctr, b_views*(1-b_ctr), size=(size))
    return sum(ra >= rb) / (1.0*size)
```

```python
fig = plt.figure(figsize=(10,5))
demonstrate(100,200, 0.04969, 0.13287, size=1000000)
```

```python
fig = plt.figure(figsize=(10,5))
demonstrate(1610, 1430, 0.04969, 0.13287, size=1000000)
```

Anscombe threshold
------------------

$$|c_a - c_b| > -\Phi^{-1} \left( \frac{n}{k+2n} \right) \sqrt(n)  $$ 

```python
duration = 20
views_per_minute = 1000

def plot_anscombe(duration, views_per_minute, serving, k_const, color):
    n = np.arange(duration*views_per_minute)
    
    if k_const:
        k = ((serving)*views_per_minute)
    else:
        k = ((duration+serving)*views_per_minute) - n
    
    anscombe = -norm.ppf((n / (k + 2.0*n))) * np.sqrt(n)
    
    pd.DataFrame(data={'upper':anscombe,'lower':-anscombe}).plot(legend=None, ax=plt.gca(), color=color)

    return plt.gca().get_legend_handles_labels()[0][-1]
```

```python
fig = plt.figure(figsize=(10,5))
h = []
l = []
servings = [20]
k_consts = [False]*3+[True]*3
colors = 'grb'*2

for s, k, color in zip(servings, k_consts, colors):
    h.append(plot_anscombe(20, 1000, s, k, color))
    l.append("serving = %d %s" % (s, "minutes"))
    plt.hold(True)
    
plt.legend(h, l)
plt.title('Anscombe stopping rule for 20 minutes of serving time')
plt.xlabel('Views')
plt.ylabel('clicks a - clicks b');
```
