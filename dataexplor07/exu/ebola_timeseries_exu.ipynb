### Use TimeSeriesModeller() to estimate a lagged model for each country

```python
class TimeSeriesModeller:
    def __init__(self):
        self.model = LinearRegression()
        
    def fit(self,y):
        max_ar_size = 10
        for ar_size in range(1, max_ar_size): # auto-regressive size
            train_start = int(len(y)/2)
            lagged = pd.concat([y.shift(i) for i in range(1,ar_size+1)], axis=1)
            self.model.fit(lagged.ix[(max_ar_size+1):train_start],y.ix[(max_ar_size+1):train_start])
            predictions = []
            for t in range(train_start, len(y)):
                prediction = self.model.predict(lagged.ix[t])
                predictions.append(prediction/y.ix[t]-1)
#                 print "prediction %s"%(prediction)
#                 print "actual %s"%(y.ix[t]-1)
            error = np.mean(predictions)
            print "model size: %s"%(ar_size)
            print "meansquarederror %f"%(error)
```

```python
Y = bfill_df.Cases_Liberia.dropna()
modeller = TimeSeriesModeller()
modeller.fit(Y)
```

The MSE results suggest an auto-regressive size of 7 is best for ebola cases in Liberia.

```python
Y = bfill_df.Deaths_Liberia.dropna()
modeller = TimeSeriesModeller()
modeller.fit(Y)
```

The MSE results suggest an auto-regressive size of 3 is best for deaths in Liberia.

### Modify the way TimeSeriesModeller() selects the training set so that it fits the model using all data except the last observed datapoint

```python
class TimeSeriesModeller2:
    def __init__(self):
        self.model = LinearRegression()
        
    def fit(self,y):
        max_ar_size = 3
        for ar_size in range(1, max_ar_size): # auto-regressive size
            train_start = int(len(y)/2) # 46
            shifted = [y.shift(i) for i in range(1,ar_size+1)]
            print "shifted: %s"%(shifted)
            lagged = pd.concat(shifted, axis=1)
            print "lagged: %s"%(lagged)
            print "selected lagged rows: %s"%(lagged.ix[(max_ar_size+1):train_start])
            # fit the model using rows (max_ar_size+1) through train_start
            self.model.fit(lagged.ix[(max_ar_size+1):train_start], y.ix[(max_ar_size+1):train_start]) # change this line
            
            predictions = []
            for t in range(train_start, len(y)):
                prediction = self.model.predict(lagged.ix[t])
                predictions.append(prediction/y.ix[t]-1)
#                 print "prediction %s"%(prediction)
#                 print "actual %s"%(y.ix[t]-1)
            error = np.mean(predictions)
            print "model size: %s"%(ar_size)
            print "meansquarederror %f"%(error)
```
