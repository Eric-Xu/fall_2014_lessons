```python
%matplotlib inline
from scipy.stats import norm
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt

tmp = pd.DataFrame({'brownian motion':np.cumsum(norm.rvs(size=100))},
                   index = pd.date_range('1/1/2000', periods=100, freq='D'))
tmp.plot()
```
