### Try k = 4

```python
km = KMeans(4,init='k-means++')
c = km.fit_predict(df_focus)
```

```python
ctroids = km.cluster_centers_
ctroids # four dimensions because df_focus has 4 predictors
```

```python
plt.plot(df_focus['GDPperCapita'][c==0], df_focus['infantMortality'][c==0],'ro')
plt.plot(df_focus['GDPperCapita'][c==1], df_focus['infantMortality'][c==1],'bo')
plt.plot(df_focus['GDPperCapita'][c==2], df_focus['infantMortality'][c==2],'go')
plt.plot(df_focus['GDPperCapita'][c==3], df_focus['infantMortality'][c==3],'yo')

plt.xlabel('GDP')
plt.ylabel('Infant Mortality Rate')
```

```python
K = range(1,10)

inertias = []
for k in K:
    # create a new KMeans object for each value of k
    kmeans = KMeans(n_clusters=k, random_state=8)
    kmeans.fit(df_focus)
    inertias.append(kmeans.inertia_)
```

```python
ideal_k = 5
# plot elbow curve
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.plot(K, inertias, 'b*-')
```

### Fit clustering results into a linear regression model

```python
X = df_focus[['GDPperCapita']] # use two brackets; otherwise will get 'tuple index out of range'
Y = df_focus[['infantMortality']]

model = LinearRegression()
model.fit(X,Y)
yhat = model.predict(X)

mean_squared_error(yhat, Y)
```
