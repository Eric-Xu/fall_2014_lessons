```python
%pylab inline
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/TeachingDataScience/data-science-course/forstudentviewing/15_kmeans/un.csv")
```

```python
focus = ['lifeMale', 'lifeFemale','infantMortality','GDPperCapita']
```

```python
df_focus = df[focus]
```

```python
##TODO: Number of missing vars for each column?
```

```python
pd.scatter_matrix(df_focus, figsize=(10,10), diagonal='kde')
```

```python
df_focus = df_focus.dropna(axis=0)
```

```python
from sklearn.cluster import KMeans
km = KMeans(3,init='k-means++')
km.fit(df_focus)
```

```python
c=km.predict(df_focus)
print km.cluster_centers_
```

```python
plt.plot(df_focus['GDPperCapita'][c==0], df_focus['lifeFemale'][c==0],'ro', label ='Undeveloped')
plt.plot(df_focus['GDPperCapita'][c==1], df_focus['lifeFemale'][c==1],'bo', label ='Developed')
plt.plot(df_focus['GDPperCapita'][c==2], df_focus['lifeFemale'][c==2],'go', label ='Developing')
plt.xlabel('GDP')
plt.ylabel('Female Life Expectancy')
legend(numpoints=1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X = df_focus[['GDPperCapita']]
Y = df_focus[['infantMortality']]

model = LinearRegression()
model.fit(X,Y)
yhat = model.predict(X)

mean_squared_error(yhat, Y)
```

```python
df_focus['c']=c
```

```python
df_focus['undeveloped']=df_focus['c'].apply(lambda x: 1 if x==0 else 0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X = df_focus[['GDPperCapita','undeveloped']]
Y = df_focus[['infantMortality']]

model = LinearRegression()
model.fit(X,Y)
yhat = model.predict(X)

mean_squared_error(yhat, Y)
```
