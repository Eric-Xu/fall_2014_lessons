```python
def make_easy_axis():
    fig = plt.figure()
    ax = fig.gca()
    patches = [Polygon(geo[0], fc='blue', alpha=0.2) for geo in coordinates]
    for patch in patches:
        ax.add_patch(patch)
    axis = [-74.008026, -73.973522, 40.710182,40.767932]
    ax.axis(axis)
    return ax
```

```python
#write helper function
ax = make_easy_axis()
ax.scatter(df.lon[100], df.lat[100], c='r', s=100)
plt.show()
```

## Understanding same day check-ins

```python
df.head()
```

```python
df['day'] = df.time.apply(lambda x: pd.datetime(x.year, x.month, x.day))
```

```python
df.head()
```

```python
df = df.sort('time')
```

```python
df.head()
```

```python
grouped = df.groupby(['user','day'])
grouped.head(10)
```

```python
#how many different users are there with more than one same day check-in?
count = 0
for name, group in grouped: # creates mini dfs
    if len(group)>4:
        print name, group
        count+=1
print count
```

```python
for name, group in grouped: # each graph is a single user on a single day
    if len(group)>3:
        ax = make_easy_axis()
        ax.scatter(group.lon[:1], group.lat[:1], c='orange', s=150) # user's first checkin of that day
        ax.scatter(group.lon, group.lat, c='r', s=60)
        ax.scatter(group.lon[-1:], group.lat[-1:], c='g', s=150) # user's last checkin
        plt.show()
```

```python
import math

avg_dist = 0.0
count = 0
for name, group in grouped:
    if len(group)>3:
        count+=1
        start = group.index[0]
        end = group.index[-1]
        start = (group.lon[start], group.lat[start])
        end = (group.lon[end], group.lat[end])
        dist = math.sqrt((start[0]-end[0])**2 + (start[1]-end[1])**2) # Pythagorian's; Euclidean Distance
        avg_dist += dist
print avg_dist, "miles"
print avg_dist/count
```
