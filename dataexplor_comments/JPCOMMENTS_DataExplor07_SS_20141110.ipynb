```python
# 3. Use TimeSeriesModeller() to estimate a lagged model for each country.
#    What about the TimeSeriesModeller class makes it not ideal for quickly changing epidemic data? 
```

```python
# Set up TimeSeriesModeller
```

```python
# Error Calculations
# i = 1 ... n
# MPE (Mean Percentage Error) = sum((Predicted(i) - Actual(i)) / Actual(i)) / n * 100%
# MSE (Mean Squared Error) = sum((Actual(i) - Predicted(i))**2) / n
# MAPE (Mean Absolute Percentage Error) = sum(abs(Actual(i) - Predicted(i)) / Actual(i))
```

```python
# The .ix attribute takes the following inputs:

# An integer or single label, e.g. 5 or 'a'
# A list or array of labels ['a', 'b', 'c'] or integers [4, 3, 0]
# A slice object with ints 1:7 or labels 'a':'f'
# A boolean array
```

```python
from sklearn.linear_model import LinearRegression 

class TimeSeriesModeller:
    
    def __init__(self):
        self.model = LinearRegression()
        
    def fit(self,dependent):
        
        # this means 1, 2, 3.. 9
        for ar_size in range(1,10):
            
            # set training data set to be the first half of the time series (minus rows 1:10)
            train_start = int(len(dependent)/2)
            
            # create lag variables up to lag 10
            lagged = pd.concat([dependent.shift(i) for i in range(1,ar_size + 1)], axis=1)
            
            # fit model where trained data starts at row 11 (index = 10) and goes to middle of the len(dependent)
            # this is a hard cutoff, irregardless of what is the ar_size of the model            
            self.model.fit(lagged.ix[10:train_start],dependent.ix[10:train_start])
            
            # create empty lists for errors
            Error_Percent = []
            Error_Percent_Abs = []
            Error_Squared = []

            for t in range(train_start, len(dependent)):
        
                # calculate predicted Y
                prediction = self.model.predict(lagged.ix[t])
                
                # calculate error, for MPE
                Error_Percent.append(((prediction - dependent.ix[t]) / dependent.ix[t]))
                
                # calculate absolute error, for MAPE
                Error_Percent_Abs.append(abs(prediction - dependent.ix[t]) / dependent.ix[t])
                
                # calculate squared error, for MSE
                Error_Squared.append((prediction - dependent.ix[t])**2)
                
            MPE = mean(Error_Percent) * 100
            MAPE = mean(Error_Percent_Abs) * 100
            MSE = mean(Error_Squared) 
        
            print "model size: %s"%(ar_size)
            print "MPE: " + str(MPE) + "%" 
            print "MAPE: " + str(MAPE) + "%"                 
            print "MSE: " + str(MSE) 
            print 
```

```python
# Run lagged model for each country
```

```python
modeller = TimeSeriesModeller()
```

```python
for column in df.columns:
    
    if column != "Day":
        
        try:
            print "results for " + column
            # dropping na values at the tail end of the df for each column
            # because not all cases and deaths in each country started at the same time
            modeller.fit(df[column].dropna())
            print 
             
        except ValueError:
            print "model for " + column + " failed."
            print "total data was " + str(len(df[column].dropna())) + " rows long"
            print "training data was " + str(len(df[column].dropna().ix[10:int(len(df[column])/2)])) + " rows long"
            print "test data was " + str(len(df[column].dropna().ix[int(len(df[column])/2):len(df[column])])) + " rows long"
            print 
```

```python
#  Q: What about the TimeSeriesModeller class makes it not ideal for quickly changing epidemic data? 
#  A: Not useful for if the dataframe is short, which is common in early epidemic data, since a large chunk of the early data
#     needs to be sacrificed for the lag.  Also, if we are splitting the dataset so that the previous 1/2 time period is used
#     to train to predict the latter half, there is no guarantee the latter half of a quickly changing epidemic dataset will
#     behave the same way as the previous half. 
```

```python
# 4. Modify the way TimeSeriesModeller() selects the training set so that it fits the model using all 
# data except the last observed datapoint. This involves changing one line of code. 
```

```python
from sklearn.linear_model import LinearRegression 

class TimeSeriesModeller:
    
    def __init__(self):
        self.model = LinearRegression()
        
    def fit(self,dependent):
        
        # this means 1, 2, 3.. 9
        for ar_size in range(1,10):
            
            # set training data set to be the entire time series, except the last data point
            train_start = int(len(dependent)-1)
            
            # create lag variables up to lag 10
            lagged = pd.concat([dependent.shift(i) for i in range(1,ar_size + 1)], axis=1)
            
            # fit model where trained data starts at row 11 (index = 10) and goes to middle of the len(dependent)
            # this is a hard cutoff, irregardless of what is the ar_size of the model
            self.model.fit(lagged.ix[10:train_start],dependent.ix[10:train_start])
            
            # create empty lists for errors
            Error_Percent = []
            Error_Percent_Abs = []
            Error_Squared = []

            for t in range(train_start, len(dependent)):
        
                # calculate predicted Y
                prediction = self.model.predict(lagged.ix[t])
                
                # calculate error, for MPE
                Error_Percent.append(((prediction - dependent.ix[t]) / dependent.ix[t]))
                
                # calculate absolute error, for MAPE
                Error_Percent_Abs.append(abs(prediction - dependent.ix[t]) / dependent.ix[t])
                
                # calculate squared error, for MSE
                Error_Squared.append((prediction - dependent.ix[t])**2)
                
            MPE = mean(Error_Percent) * 100
            MAPE = mean(Error_Percent_Abs) * 100
            MSE = mean(Error_Squared) 
        
            print "model size: %s"%(ar_size)
            print "MPE: " + str(MPE) + "%" 
            print "MAPE: " + str(MAPE) + "%"                 
            print "MSE: " + str(MSE) 
            print 
```

```python
# Q: What are the tradeoffs with this approach? 
# A: You train on more data, over a longer time, so the accuracy of the model improves.  
#    Also, no need to worry that, after taking out the training data, there is nothing left
#    But since the testing dataset is only one data point, it's hit or miss for model accuracy output.  
```
