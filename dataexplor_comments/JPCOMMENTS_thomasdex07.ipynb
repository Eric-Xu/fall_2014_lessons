```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import Ebola Data #
df = pd.DataFrame.from_csv('https://raw.githubusercontent.com/cmrivers/ebola/master/country_timeseries.csv',
                 index_col=0)
```

```python
#reproduce epidemic curves for each county

df = df.fillna(method = 'bfill') # will fill data from first data point
# df = df.interpolate() # will fill data before data points

# The data from the most recent day
mostrecent = df[0:1]
```

```python
# Split data into cases and deaths
cases_column = []
deaths_column = []
```

```python
for column in df.columns:
    head = column.split('_')
    if head[0] == 'Cases':
        cases_column.append(column)
    if head[0] == 'Deaths':
        deaths_column.append(column)
```

```python
# Plot Cases and Deaths
cases = df[cases_column]
cases.plot()
show()

deaths = df[deaths_column]
deaths.plot()
show()
```

```python
# Lag Corr Plots

def acf(df):
    import seaborn as sns
    lagged = pd.concat([df.shift(i) for i in range(10)], axis=1)
    lagged.columns = ["lag%s"%i for i in range(10)]
    sns.corrplot(lagged)
```

```python
for c in deaths.columns:
    print c
    acf(deaths[c])
    show()
```

```python
for c in cases.columns:
    print c
    acf(cases[c])
    show()
```
