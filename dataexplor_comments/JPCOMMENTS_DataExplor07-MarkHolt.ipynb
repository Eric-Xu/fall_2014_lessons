```python
class TimeSeriesModeller:
    def __init__(self):
        self.model = LinearRegression()
        
    def fit(self,y):
        #remove this from the inner loop - only need to calculate it once
        #train_start = int(len(y)/2)
        train_start = len(y)-1
        
        for ar_size in range(1,10):
            
            #the initial range was from 11 to train_start, but can begin at ar_size
            lagged = pd.concat([y.shift(i+1) for i in range(0,ar_size)], axis=1)
            self.model.fit(lagged.ix[ar_size:train_start],y.ix[ar_size:train_start])
            predictions = []
            for t in range(train_start, len(y)):
                prediction = self.model.predict(lagged.ix[t])
                #Watch out for divide by zero
                #print "prediction = ", prediction
                #print "actual = ", y.ix[t]
                #if(y.ix[t] != 0):
                #    predictions.append(prediction/y.ix[t]-1)
                #    if(prediction == 0):
                #       predictions.append(1-1)
                #I do not understand the above mse calculation. The division by y.ix[t] causes all sorts of problems???
                sqerr = (prediction - y.ix[t]) * (prediction - y.ix[t])
                predictions.append(sqerr)
            error = np.mean(predictions)
            print "model size: %d : MSE = %0.10f"%(ar_size, error)
            #print "meansquarederror %f"%(error)
```

##We are using more data to train on, but we are also using older data. This explains the worsening in the MSE. This is particularly noticeable for this data which is rapidly changing. So older data will be less relevant and useful to predict the next data point

##Need to define a separate log function to handle zeros in the data
##Reinforces the wisdom on NOT using "interpolate" to deduce missing values!

```python
def performLog(x):
    if x!=0:
        return np.log10(x)
    else:
        return x
```

```python
deaths_df_lg = deaths_df.applymap(performLog)
```

```python
cases_df_lg = cases_df.applymap(performLog)
```
