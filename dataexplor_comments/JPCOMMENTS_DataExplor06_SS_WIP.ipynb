```python
from bs4 import BeautifulSoup
from urllib2 import urlopen
import pandas as pd
import numpy as np
```

```python
### Step 1: On the master page (http://www.cs.cmu.edu/~./gdead/setlists.html), grab the url for each year
```

```python
url_year = 'http://www.cs.cmu.edu/~./gdead/setlists.html'
page_year = urlopen(url_year)
soup_year = BeautifulSoup(page_year)

# print soup_year.prettify()
# print soup_year.get_text()
```

```python
# want the end result to be like this: 
# [(Year 1, Year URL 1), (Year 2, Year URL 2) ...]
year_id_1 = []
year_url = []

for link_year in soup_year.find_all('a'):
    if link_year.get('href')[-4:] == 'html':
        # print 'http://www.cs.cmu.edu/~./gdead/' + link_year.get('href'), link_year.contents[0]
        year_id_1.append(link_year.contents[0])
        year_url.append('http://www.cs.cmu.edu/~./gdead/' + link_year.get('href'))

# zip together to make a tuple of the year, and that year's url
zip_1 = zip(year_id_1, year_url)

# visual check first 5: 
for year in zip_1[:5]:
    print year[0], year[1]

# create dataframe from the tuple:
df_1_final = pd.DataFrame(zip_1, columns=['year_id_1', 'year_url'])
df_1_final.head()
```

```python
### Step 2: For each year, find all concert names and all set information for each concert
```

```python
# initially, want the result to be like this:
# [(Year 1, Concert Name 1, Concert Set List URL 1), (Year 2, Concert Name 2, Concert Set List URL 2) ...]

year_id_2 = []
set_url = []
concert_name = []

for url in year_url:
    for link_set in BeautifulSoup(urlopen(url)).find_all('a'):
        if link_set.get('href')[-3:] == 'txt' and link_set.get('href') != 'dead-sets/93/useless-deadstats.txt':
            year_id_2.append('19' + link_set.get('href')[-6:-4]) 
            set_url.append('http://www.cs.cmu.edu/~./gdead/' + link_set.get('href'))
            concert_name.append(link_set.contents[0])
    
# zip together to make a tuple of the year, and that year's url
zip_2 = zip(year_id_2, set_url, concert_name)

# change tuple into dataframe
df_2_raw = pd.DataFrame(zip_2, columns=['year_id_2', 'set_url', 'concert_name'])
df_2_raw.head()
```
